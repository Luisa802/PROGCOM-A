```python
comida=["combo 1", "combo 2", "combo 3", "combo 4", "Papas", "jugos naturales", "cerveza", "agua", "Gaseosa" ]
precios=[20000,25000,18000,24500,3000,7000,3500,4000,2500]
```

```python
for menu in range(len(comida)):
  print(comida[menu], "-", precios[menu])
```

```python
a=0
r="s"
user="s"
while user=="s":
  while r=="s":
    x=input("Qué desea ordenar?: ")
    if x=="combo 1": a=a+20000
    elif x=="combo 2": a=a+25000
    elif x=="combo 3": a=a+18000
    elif x=="combo 4": a=a+24500
    elif x=="Papas": a=a+3000
    elif x=="jugos naturales": a=a+7000
    elif x=="Cerveza": a=a+3500
    elif x=="Agua": a=a+4000
    elif x=="Gaseosa": a=a+2500
    else:
      x=input("Desea algo más?")
      if x=="s": continue
      else: break
  user=input("Hay más usuarios?: ")
print(f"Usted ha ganado ${a} en el día")
```

```python
a=0
r="si"
user="si"
while user=="si":
  while r=="si":
    x=input("Qué desea ordenar?: ")
    if x in comida:
      cantidad = int(input("¿Cuántas unidades desea ordenar?: "))
      a = a + precios[comida.index(x)] * cantidad
    else:
      print("No esta ese producto")
    while True:
      x=input("Desea algo más? (si/no): ").lower()
      if x in ["si", "no"]:
        break
      else:
        print("Por favor, ingrese 'si' o 'no'.")

    if x=="si": continue
    else: break
  user=input("Hay más usuarios?: ").lower()
print(f"Usted ha ganado ${a} en el día")
```
