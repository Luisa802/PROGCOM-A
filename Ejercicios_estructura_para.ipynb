# 18	Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

```python
#forma 1
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1):
    factorial = factorial * i

if numero < 0:
    print("No existen factoriales para números negativos")
elif numero == 0:
    print("El factorial de 0 es 1")
else:
    print("El factorial de", numero, "es:", factorial)
```

```python
#forma 2
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1):
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
```

```python
#forma 3
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1, 1):   # VI=1, VF=numero, P=1
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
```

```python
#codigo del ejercicio en java forma 3
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Escribe un número para sumar sus dígitos: ");
        String numero = sc.nextLine();

        int suma = 0;

        for (int i = 0; i < numero.length(); i++) {
            char digito = numero.charAt(i);
            suma = suma + Character.getNumericValue(digito);
        }

        System.out.println("Resultado: " + suma);

        sc.close();
    }
}
```

#23. Escribe un programa que determine si un número es primo.

```python
#froma 1
numero = int(input("Ingrese un número: "))

for i in range(numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif i >= 2 and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
```

```python
#forma 2
numero = int(input("Ingrese un número: "))

for i in range(2, numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif numero % i == 0:
        print("No es un número primo")
        break
else:
    if numero > 1:
        print("Es un número primo")
```

```python
#forma 3
numero = int(input("Ingrese un número: "))

for i in range(1, numero + 1, 1):
    if numero <= 1:
        print("No es un número primo")
        break
    elif i > 1 and i < numero and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
```

Ejercicio en java forma 1

image.png

```python
#codigo del ejercicio en java forma 1
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);
        System.out.print("Ingrese un número: ");
        int numero = sc.nextInt();

        if (numero <= 1) {
            System.out.println("No es un número primo");
        } else {

            boolean esPrimo = true;

            for (int i = 2; i < numero; i++) {
                if (numero % i == 0) {
                    esPrimo = false;
                    break;
                }
            }

            if (esPrimo) {
                System.out.println("Es un número primo");
            } else {
                System.out.println("No es un número primo");
            }
        }

        sc.close();
    }
}
```

```python
#codigo del ejercicio en java de la forma 1
import java.util.Scanner;

public class potencia {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese la base: ");
        int base = sc.nextInt();

        System.out.print("Ingrese el exponente: ");
        int exponente = sc.nextInt();

        int resultado = 1;

        for (int i = 0; i < exponente; i++) {
            resultado = resultado * base;
        }

        System.out.println("La potencia es: " + resultado);
    }
}
```
