#ESTRUCTURA WHILE - MIENTRAS

Sintaxis en python
````
while (condición inicial(CI)):
  parte del código a repetir
  Modificador de condición inicial (CI)
````
CI: la CI debe tener un valor de verdad VERDADERO
(TRUE) para poder ingresar al while.


Sintaxis en java
````
while (CI){
  código que debe repetirse
  Un modificador de CI
}
````

`````python
yo = 0
while yo < 1 or yo > 3:
    entrada = input("Elije entre 1. 🪨, 2. 📜, 3. ✂️: ")

    try:
        yo = int(entrada)
        if yo < 1 or yo > 3:
            print("Ese valor no es válido. Debe ser 1, 2 o 3.")
    except ValueError:
        print("¡Error! Por favor, introduce un número, no letras.")
        yo = 0
print(f"Seleccionaste la opción: {yo}")
`````
