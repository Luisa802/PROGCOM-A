#16. Crea un programa que muestre la tabla de multiplicar de un número hasta el 10.

```python
#programa que muestre la tabla de multiplicar de un número hasta el 10.
numero_tabla = int(input("¿De qué número quieres la tabla?: "))
i = 1
while (i <= 10):

    resultado = numero_tabla * i
    print(f"{numero_tabla} x {i} = {resultado}")
    i = i + 1
```

# 18	Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

```python
numero = int(input("Ingresa un número para calcular su factorial: "))


while numero < 0:
    print("Error, No existen factoriales negativos.")
    numero = int(input("Ingresa un número válido (0 o positivo): "))

factorial = 1
ci = numero

while ci > 0:
    factorial = factorial * ci
    ci = ci - 1

print(f"El factorial de {numero} es: {factorial}")
```

Ejercicio en java


image.png

```python
#Codigo del ejercicio en java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Ingresa un número para calcular su factorial: ");
        int numero = sc.nextInt();


        while (numero < 0) {
            System.out.println("Error, No existen factoriales negativos.");
            System.out.print("Ingresa un número válido (0 o positivo): ");
            numero = sc.nextInt();
        }

        long factorial = 1;
        int ci = numero;

        while (ci > 0) {
            factorial = factorial * ci;
            ci = ci - 1;
        }

        System.out.println("El factorial de " + numero + " es: " + factorial);

        sc.close();
    }
}
```

#19.Crea un programa que muestre los primeros n números de la serie Fibonacci.

```python
n=int(input("Ingresa un numero:"))

while n <= 0:
    print("Error, Debes ingresar un número mayor a 0.")
    n = int(input("Ingresa un número válido: "))

a = 0
b = 1
contador = 1

print("Serie Fibonacci:")

if n == 1:
    print(a)
else:
    while contador <= n:
        print(a)
        siguiente = a + b
        a = b
        b = siguiente
        contador = contador + 1
```

Ejercicio en java


image.png

```python
#codigo del ejercicio en java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Ingresa un numero: ");
        int n = sc.nextInt();

        while (n <= 0) {
            System.out.println("Error, Debes ingresar un número mayor a 0.");
            System.out.print("Ingresa un número válido: ");
            n = sc.nextInt();
        }

        long a = 0;
        long b = 1;
        int contador = 1;

        System.out.println("Serie Fibonacci:");

        if (n == 1) {
            System.out.println(a);
        } else {
            while (contador <= n) {
                System.out.print(a + " ");
                long siguiente = a + b;
                a = b;
                b = siguiente;
                contador = contador + 1;
            }
        }

        sc.close();
    }
}
```

#20. Escribe un programa que adivine entre un conjunto de diez colores definidos por el usuario, con un máximo de 10 intentos.

```python
import random

colores = []

for i in range(10):
    color = input("Ingrese un color: ")
    colores.append(color)

color_secreto = random.choice(colores)

intentos = 0

while intentos < 10:
    intento = input("Adivina el color 🔮: ")

    if intento == color_secreto:
        print("¡Adivinaste el color!🎊")
        break
    else:
        print("Color incorrecto😬")
        intentos += 1

if intentos == 10:
    print("Se acabaron los intentos 😔. El color era:", color_secreto)
```

```python
#Codigo del ejercicio en java
import java.util.Scanner;
import java.util.Random;

public class AdivinarColor {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);
        Random random = new Random();

        String[] colores = new String[10];

        for(int i = 0; i < 10; i++){
            System.out.print("Ingrese un color: ");
            colores[i] = sc.nextLine();
        }

        String colorSecreto = colores[random.nextInt(10)];

        int intentos = 0;

        while(intentos < 10){

            System.out.print("Adivina el color 🔮: ");
            String intento = sc.nextLine();

            if(intento.equals(colorSecreto)){
                System.out.println("¡Adivinaste el color! 🎊");
                break;
            }else{
                System.out.println("Color incorrecto 😬");
                intentos++;
            }
        }

        if(intentos == 10){
            System.out.println("Se acabaron los intentos 😔. El color era: " + colorSecreto);
        }

    }
}
```

#22. Crea un programa que calcule la suma de los dígitos de un número.

```python
numero = input("Introduce un número entero: ")
suma = 0
posicion = 0
while (posicion < len(numero)):
    digito = int(numero[posicion])
    suma = suma + digito
    posicion = posicion + 1
print("La suma de los dígitos es:", suma)
```

Ejercicio en java


image.png

```python
#codigo del ejercicio en java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Introduce un número entero: ");
        String numero = sc.next();

        int suma = 0;
        int posicion = 0;


        while (posicion < numero.length()) {

            int digito = Character.getNumericValue(numero.charAt(posicion));
            suma = suma + digito;
            posicion = posicion + 1;
        }

        System.out.println("La suma de los dígitos es: " + suma);
        sc.close();
    }
}
```

#23. Escribe un programa que determine si un número es primo.

```python
numero=int(input("Ingresa un numero: "))

while numero < 2:
    print("Error, Los números primos son mayores a 1.")
    numero = int(input("Ingresa un número válido (mayor a 1): "))

divisor = 2
es_primo = True

while divisor <= numero - 1:
    if numero % divisor == 0:
        es_primo = False
        divisor = numero
    divisor = divisor + 1

if es_primo:
    print(numero, "ES un número primo")
else:
    print(numero, "NO es un número primo")
```

Ejercicio en java


image.png

```python
#Codigo del ejercicio en java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Ingresa un numero: ");
        int numero = sc.nextInt();

        while (numero < 2) {
            System.out.println("Error, Los números primos son mayores a 1.");
            System.out.print("Ingresa un número válido (mayor a 1): ");
            numero = sc.nextInt();
        }

        int divisor = 2;
        boolean esPrimo = true;

        while (divisor <= numero - 1) {
            if (numero % divisor == 0) {
                esPrimo = false;
                divisor = numero;
            }
            divisor = divisor + 1;
        }

        if (esPrimo) {
            System.out.println(numero + " ES un número primo");
        } else {
            System.out.println(numero + " NO es un número primo");
        }
        sc.close();
    }
}
```

#24. Desarrolla un programa que muestre el patrón de un triángulo con asteriscos.

```python
altura = int(input("Ingrese la altura del triángulo 📐: "))

fila = 1

while fila <= altura:
    print("*" * fila)
    fila += 1
```

Ejercicio en java


image.png

```python
#Codigo del ejercicio en java
import java.util.Scanner;

public class Triangulo {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese la altura del triángulo 📐 : ");
        int altura = sc.nextInt();

        int fila = 1;

        while(fila <= altura){

            int contador = 0;

            while(contador < fila){
                System.out.print("*");
                contador++;
            }

            System.out.println();
            fila++;
        }
    }
}
```

#25. Crea un programa que calcule la potencia de un número sin usar el operador **.

```python
base = int(input("Ingrese la base: "))
exponente = int(input("Ingrese el exponente: "))

resultado = 1
contador = 0

while contador < exponente:
    resultado = resultado * base
    contador += 1

print("La potencia es:", resultado)
```

```python
#Codigo del ejercicio en java
import java.util.Scanner;

public class Potencia {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese la base: ");
        int base = sc.nextInt();

        System.out.print("Ingrese el exponente: ");
        int exponente = sc.nextInt();

        int resultado = 1;
        int contador = 0;

        while(contador < exponente){
            resultado = resultado * base;
            contador++;
        }

        System.out.println("La potencia es: " + resultado);

    }
}
```
