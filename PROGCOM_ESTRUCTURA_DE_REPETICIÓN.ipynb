##1. PARA - FOR

* Se emplea cuando se sabe cuántas veces se repetirá la acción.
El conteo siempre inicia desde**0**

Sintaxis general
````
for (variable) in (algo):
 La acción que se va a repertir

#FORMA2
for variable in range(cantidad de veces):
 La acción que se va a repetir.

#FORMA3
for variable in range(VI,VF+1,P):
 La acción que se va arepetir.
````

Donde:

 *VI: Valor inicial (dónde inicio)

 *VF: Valor final (dónde finalizo)

 *P: Paso. Casa cuánto cuento

`````python
#Ejercicio imprimir los numeros pares del 1 al 10
#FORMA 1
for num in range(11):
  if num%2==0: print(num)
`````

`````python
#FORMA 2
for num in range(1,11):
  if num%2==0: print(num)
`````

`````python
#FORMA 3
for num in range(0,11,2):
  print(num)
`````
