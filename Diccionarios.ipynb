#DICCIONARIO
1. Tipo de dato como las listas. **Si son mutables**.
2. Se utilizan para almacenar datos en pares, clave:valor. La clave **Debe ser alfanumérica**.
````
variablre={clave:valor}
````
3. Los diccionarios son coleccones ordenadas, modificadas y **No permite duplicados**.


`````python
estudiante={"nombre": "Luisa",
            "ID": "U00186483",
            "Edad": "18",
            "Carrera": "Ingenieria Industrial"}
print(estudiante)
`````

`````python
#Listas
galletas=["angelitos","diablitos","dinosaurio"]
galletas_precio=[20000,15000,15000]
galletas_cant=[3,5,10]
`````

`````python
def menu_galletas():
  for i in range(3):
    print(f"{galletas[i]}:{galletas_precio[i]}, cantidad:{galletas_cant[i]}")
`````

`````python
menu_galletas()
`````

`````python
#Diccionarios
galleta={"tipo de galleta":["angelitos","diablitos","dinosaurios"], "precio":[20000,15000,25000], "cantidad":[3,5,10]}
`````

`````python
def menu_galleta_dict():
  for i in range(3):
    print(f"{galleta["tipo de galleta"][i]}:{galleta["precio"][i]}, cantidad:{galleta["cantidad"][i]}")
`````

`````python
menu_galleta_dict()
`````
