#Ejercicio 5
Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

```python
estudiantes = {
    "Ana": 4.5,
    "Luis": 3.8,
    "Sofia": 4.2
}

def promedio_general(dic):
    suma = sum(dic.values())
    return suma / len(dic)

def promedio_estudiante(dic, nombre):
    return dic[nombre]

print("Promedio general:", promedio_general(estudiantes))
print("Nota de Ana:", promedio_estudiante(estudiantes, "Ana"))
```

#Ejercicio 7
Dada la palabra "magia", cuenta cuántas veces aparece cada letra usando un diccionario.

```python
palabra = "magia"
conteo = {}

for letra in palabra:
    if letra in conteo:
        conteo[letra] += 1
    else:
        conteo[letra] = 1

print(conteo)
```

#Ejercicio 8
Dada una lista de elementos recolectados en una expedición, cuenta cuántas veces aparece cada elemento.

Lista sugerida: ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]

```python
lista = ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo = {}

for item in lista:
    if item in conteo:
        conteo[item] += 1
    else:
        conteo[item] = 1

print(conteo)
```
